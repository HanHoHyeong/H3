# **캘리포니아 주택 위치**


````python
# prompt: /content/sample_data/california_housing_train.csv 로부터 latitude와 latitude 를 한눈에 보게 그래프로 정리해줘

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Create the scatter plot
plt.figure(figsize=(10, 6))  # Adjust figure size for better visibility
plt.scatter(df['longitude'], df['latitude'], alpha=0.4, s=df['population'] / 100) # Added population as size
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('California Housing Locations')
plt.colorbar(label='Population') # Show population scale

plt.show()
````

# **주택 중간 연령**

````python
# prompt: /content/sample_data/california_housing_train.csv 를 이용해서 housing_median_age를 한눈에 볼 수 있게 그래프로 만들어줘

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Create a histogram of housing_median_age
plt.figure(figsize=(10, 6))
plt.hist(df['housing_median_age'], bins=20, edgecolor='black')  # Adjust number of bins as needed
plt.xlabel('Housing Median Age')
plt.ylabel('Frequency')
plt.title('Distribution of Housing Median Age in California')
plt.grid(axis='y', alpha=0.75)
plt.show()
````

````python
# prompt: /content/sample_data/california_housing_train.csv 를 이용해서 total_rooms와 total_bedrooms를 보기쉽게 그래프로 보여줘 bedrooms와 rooms는 색깔을 다르게 하는데 뚜렷하게 볼 수 있는 색깔로 해줘

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Create the scatter plot with different colors for total_rooms and total_bedrooms
plt.figure(figsize=(10, 6))
plt.scatter(df['total_rooms'], df['total_bedrooms'], c='red', label='Total Rooms', alpha=0.5) # Red for total_rooms
plt.scatter(df['total_rooms'], df['total_bedrooms'], c='blue', label='Total Bedrooms', alpha=0.5) # Blue for total_bedrooms


plt.xlabel('Total Rooms')
plt.ylabel('Total Bedrooms')
plt.title('Total Rooms vs. Total Bedrooms')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# prompt: /content/sample_data/california_housing_train.csv를 이용해서 세계지도모양을 사용해서 longitud와 latitude를 해당되면 진하게 그려줘

import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

# Load the dataset
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Create the map using plotly express
fig = px.scatter_mapbox(df, lat="latitude", lon="longitude", hover_name="population",
                        color_discrete_sequence=["fuchsia"], zoom=5, height=600)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
````
